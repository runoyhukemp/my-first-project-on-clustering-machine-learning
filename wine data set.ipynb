# data clustering

```python
from sklearn.preprocessing import normalize
scaled=normalize(df)
scaled=pd.DataFrame(scaled)
scaled.head()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10,7))
dendrogram =sch.dendrogram(sch.linkage(scaled,method="ward"))
plt.title('Dendrogram')
plt.xlabel('Quality of wine')
plt.ylabel('apply all features')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
# cluster.fit_predict(datasubset)
pred=cluster.fit_predict(scaled)
pred
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled,pred)
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("E:/DATA SCIENCE/jan/winequality_red.csv")
df
```

```python
df.head()
```

```python
z=df.columns
z
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms
```

```python
df=ms.fit_transform(df)
df
```

```python
df=pd.DataFrame(df)
df
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2,random_state=0)
km.fit(df)
```

```python
sse=[]
k_rng=range(1,5)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squarred error')
plt.plot(k_rng,sse)
```

```python
km.cluster_centers_
```

```python
pred=km.predict(df)
pred
```

```python
km.inertia_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df,pred)
```
